## Exploratory Data Analysis (EDA)

```python
# Pertanyaan Bisnis 1: Lima kota mana yang memiliki jumlah pelanggan terbanyak?
city_counts = customers_dataset['customer_city'].value_counts().nlargest(5)
print("\nLima Kota Teratas:")
print(city_counts)
```

```python
# Pertanyaan Bisnis 2: Provinsi mana yang memiliki pelanggan terbanyak?
state_counts = customers_dataset['customer_state'].value_counts()
print("\nJumlah Pelanggan Berdasarkan Provinsi:")
print(state_counts)
```

# 6. Visualization & Explanatory Analysis

```python
# Visualisasi untuk Pertanyaan 1: Pie Chart untuk Lima Kota Teratas
plt.figure(figsize=(8, 8))
plt.pie(city_counts, labels=city_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Persentase Pelanggan di 5 Kota Teratas')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Interpretasi Pertanyaan 1
print("Interpretasi Pertanyaan 1:")
for city, count in zip(city_counts.index, city_counts.values):
    print(f"Kota: {city}, Jumlah Pelanggan: {count}, Persentase: {count / city_counts.sum() * 100:.2f}%")
```

```python
# Visualisasi untuk Pertanyaan 2: Bar Chart untuk Jumlah Pelanggan Berdasarkan Provinsi
plt.figure(figsize=(12, 6))
sns.barplot(x=state_counts.index, y=state_counts.values, palette='viridis')
plt.title('Jumlah Pelanggan Berdasarkan Provinsi')
plt.xlabel('Provinsi')
plt.ylabel('Jumlah Pelanggan')
plt.xticks(rotation=45)

# Menambahkan anotasi jumlah di atas setiap batang
for i in range(len(state_counts)):
    plt.text(i, state_counts.values[i] + 5, state_counts.values[i], ha='center')

plt.show()
```

```python
# Interpretasi Pertanyaan 2
print("Interpretasi Pertanyaan 2:")
for state, count in zip(state_counts.index, state_counts.values):
    print(f"Provinsi: {state}, Jumlah Pelanggan: {count}")
```

```python
# 7. Analisis Lanjutan
# Analisis demografis atau perilaku lebih lanjut dapat dilakukan berdasarkan data yang ada
demographic_analysis = customers_dataset.groupby('customer_state').agg({
    'customer_id': 'count',
    'customer_zip_code_prefix': 'nunique'
}).reset_index()
print("\nAnalisis Demografis per Provinsi:")
print(demographic_analysis)
```

##Analisis Lanjutan

```python
#Geospatial Analysis
# Menghitung jumlah pelanggan per kode pos
zip_code_counts = customers_dataset['customer_zip_code_prefix'].value_counts()

# Visualisasi kode pos dengan jumlah pelanggan terbanyak (Top 10)
plt.figure(figsize=(12, 6))
sns.barplot(x=zip_code_counts.nlargest(10).index.astype(str), y=zip_code_counts.nlargest(10).values, palette='viridis')
plt.title('Jumlah Pelanggan Berdasarkan Kode Pos (Top 10)')
plt.xlabel('Kode Pos')
plt.ylabel('Jumlah Pelanggan')
plt.xticks(rotation=45)

# Menambahkan anotasi jumlah di atas setiap batang
for i in range(len(zip_code_counts.nlargest(10))):
    plt.text(i, zip_code_counts.nlargest(10).values[i] + 5, zip_code_counts.nlargest(10).values[i], ha='center')

plt.show()

# Interpretasi hasilnya
print("\nInterpretasi Geospatial:")
for zip_code, count in zip(zip_code_counts.nlargest(10).index, zip_code_counts.nlargest(10).values):
    print(f"Kode Pos: {zip_code}, Jumlah Pelanggan: {count}")
```

```python
#Clustering
# Manual Grouping berdasarkan provinsi, misalnya, pelanggan dari provinsi tertentu bisa dikategorikan sebagai prioritas
def categorize_state(state):
    if state in ['SP', 'RJ']:  # Misalkan SP dan RJ adalah provinsi dengan prioritas tinggi
        return 'High Priority'
    elif state in ['MG', 'RS']:  # Provinsi dengan prioritas menengah
        return 'Medium Priority'
    else:
        return 'Low Priority'

customers_dataset['Customer_Group'] = customers_dataset['customer_state'].apply(categorize_state)
print(customers_dataset[['customer_id', 'customer_state', 'Customer_Group']].head())
```

##Menyimpan Model

```python
import pickle
```

```python
#Menyimpan hasil clustering 'Customer_Group' sebagai model
model = customers_dataset[['customer_id', 'Customer_Group']]

# Menyimpan model ke file
filename = 'CustomerDataset_model.sav'
pickle.dump(model, open(filename, 'wb'))
```
